### 7. (3đ) Tạo pie chart thể hiện phần trăm của cột race và sex (sẽ có 5*2=10 categories tổng thể trong 1 hình vẽ). Nhận xét xem nhóm nào chiếm tỉ lệ nhiều nhất.

```python
# Your code
race = df['race'].values
print(race)
# race = [1 if i == 'White'
sex = df['sex'].values
print(sex)

dicttion = {
    " Male": {
        " White": 0,
        " Black": 0,
        " Asian-Pac-Islander": 0,
        " Amer-Indian-Eskimo": 0,
        " Other": 0
    },
    " Female": {
        " White": 0,
        " Black": 0,
        " Asian-Pac-Islander": 0,
        " Amer-Indian-Eskimo": 0,
        " Other": 0
    }
}


for i in range(len(race)):
    dicttion[sex[i]][race[i]] += 1

```

```python
print(dicttion)

arr = [i for i in dicttion[" Male"].values()]
arr.extend([i for i in dicttion[" Female"].values()])
arr

plt.pie(arr, labels=['White', 'Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other', 'White', 'Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other'], autopct='%1.1f%%', startangle=140)
```

### 20. (6đ) Tỉ lệ người married hay single men (marital-status feature) chiếm nhiều hơn trong số những người kiếm được >50k? Những người có tình trạng hôn nhân (married) có marital-status feature bắt đầu với Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), còn lại được coi là single men.

```python
# You code here
```

```python
more_50k = df[df["salary"] == " >50K"]

married_sum = 0
single_sum = 0
for i in [" Married-civ-spouse", " Married-AF-spouse", " Married-spouse-absent"]:
    married_sum += more_50k[more_50k["marital-status"] == i].shape[0]

for i in [" Never-married", " Divorced", " Separated", " Widowed"]:
    single_sum += more_50k[more_50k["marital-status"] == i].shape[0]

print(married_sum)
print(single_sum)
```

### 23. (10đ) Số người da trắng (white in race feature) kiếm được số tiền >50k ở nước Mỹ chiếm bao nhiêu phần trăm. Câu hỏi tương tự với người da đen (black in race feature).
